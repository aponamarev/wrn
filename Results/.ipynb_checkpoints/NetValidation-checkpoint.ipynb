# 2. Create supporting functions

- normalize(img) provides a normalized version of the image with per channel mean shift to 0 and std correction to 1
- load_img_from_desc(desc, path, index) - takes a descriptor, path to where the images are stored, and an index of the image. The output of the function is an image that will be displayed before the image is returned.
- provide_top5_predictions(img, net) - takes a forward pass of the img through the net and returns sorted top5 results
- provide_simial5(group_desc, path, id) - displays 5 images of the same class_id

```lua
function normalize(img)
    --Normalizes mean and standard deviation of the image to 0 and 1 respectively
    img = img:float()
    assert(img:dim()==3, "Error: Wrong number of dimentions. Image expected with c*h*w dimentions. Instead " .. img:dim() .. " tensor was provided.")
    local flat_img = torch.view(img, 3, img:size(2) * img:size(3)):float()
    local m = torch.mean(flat_img, 2):squeeze()
    local std = torch.std(flat_img, 2):squeeze()
    for i=1,3 do
        img[i]:add(-m[i])
        img[i]:div(std[i])
    end
    return img
end
--
function load_img(path, img_name)
    local path = paths.concat(path, img_name)
    assert(paths.filep(path), "Error: File doesn't exit at path: " .. path)
    local img = image.load(path)
    img = image.scale(img, img_dim, img_dim)
    img = normalize(img)
    display_image(img)
    img = torch.view(img, 1, 3, img_dim, img_dim)
    return img:float()
end
--
function load_img_from_desc(desc, path, index)
    local img_name = desc.file_name[index]
    local age = desc.age[index]
    local identity_id = desc.identity_id[index]
    local img_class = desc.classes[identity_id]
    print("img_name: " .. img_name .. ", name: " .. img_class .. ", age: " .. age .. ", id: " .. identity_id)
    return load_img(path, img_name), identity_id, img_class, age
end
--
function provide_top5_predictions(img, _net, desc)
    local topX = 5
    local output = _net:forward(img):squeeze()
    local prob, id = torch.sort(output, 1, true)
    id = id[{{1, topX}}]
    prob = prob[{{1, topX}}]
    print("Top5:")
    local names = {}
    for i =1, id:size(1) do
        local name = desc.classes[id[i]]
        print(id[i], string.format('%.2f percent',prob[i]*100), name)
        names[i] = name
    end
    return id, prob, names
end
--
function provide_simial5(group_desc, path, id)
    assert(paths.dirp(path), "Error: Provided path (".. path .. ") doesn't exist!")
    local topX = 5
    local topResults = {images = {}, age = {}}
    for i = 1, topX do
        local p_ = paths.concat(path, group_desc[id].file_name[i])
        print(p_)
        topResults.images[i] = image.load(p_)
        topResults.age[i] = image.load(group_desc[id].age[i])
    end
    print(group_desc[id].class)
    print(topResults.age[1], topResults.age[2], topResults.age[3], topResults.age[4], topResults.age[5])
    display_image(topResults.images[1],topResults.images[2],topResults.images[3],topResults.images[4],topResults.images[5])
end
```

## Launch the classification ##

- load image
- run top 5 prediction

```lua
local descriptor_used = val_desc
anchor_img, ground_trueth_id, ground_trueth_class, train_age = load_img_from_desc(descriptor_used, cacd_path, 1334)
predicted_id, prediction_prob, predicted_names = provide_top5_predictions(anchor_img, net, descriptor_used)
provide_simial5(group_descriptor, cacd_path, ground_trueth_id)
```
